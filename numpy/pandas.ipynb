```python
import datetime
import matplotlib.pyplot as plt
```

```python
def read_weather(filename):
    parser = lambda date: datetime.datetime.strptime(date, '%d.%m.%Y %H:%M')
    df = pd.read_csv(filename, sep=';', comment='#', index_col=False,
                     parse_dates=[0,], date_parser=parser)
    df.rename(columns={df.columns[0] : 'time'}, inplace=True)
    return df
```

```python
df_time1 = read_weather("UUDD.csv.gz")
df_time1.head()
```

```python
df_time2 = read_weather("UUEE.csv.gz")
df_time2.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_time1['time'], df_time1['T'], '.', alpha=0.5, label="UUDD")
plt.plot(df_time2['time'], df_time2['T'], '.', alpha=0.5, label="UUEE")
plt.legend()
```

### Солнечная активность

```python
#
# https://observethesun.com/indices?current=2021-02-16&objects=3f&past=1970-01-01
#
with open("sun_indices_spot.json", encoding="utf-8-sig") as f:
    x = json.load(f)
    df_time1 = pd.DataFrame(x["data"])
    df_time1.rename(columns={df_time1.columns[0] : 'time'}, inplace=True)
    df_time1.time = pd.to_datetime(df_time1.time, format='%Y-%m') + pd.Timedelta('14 day')
    df_time1.set_index("time", inplace=True)
    for column in df_time1.columns:
        df_time1[column] = pd.to_numeric(df_time1[column])
        
df_time1.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_time1['wolf'])
```

```python
from scipy.io import netcdf
```

```python
with netcdf.netcdf_file('MSR-2.nc') as netcdf_file:
    data = netcdf_file.variables
    ozon = data['Average_O3_column'].data
    lat  = data['latitude'].data
    long = data['longitude'].data
    time = data['time'].data
```

```python
fig = plt.figure(figsize=(15,6))

plt.subplot(111, projection="hammer")
plt.contourf(*(np.meshgrid(long / 180 * np.pi, lat / 180 * np.pi)), ozon[0], cmap='gray_r')
plt.colorbar()
plt.grid()
```

```python
def average_eq(limit_lat):
    s, e = np.searchsorted(lat, np.asarray([-limit_lat, limit_lat]))
    return ozon[:, s:e, :].mean(axis=(1,2))
```

```python
def average_polar(limit_lat):
    s, e = np.searchsorted(lat, np.asarray([-limit_lat, limit_lat]))
    return np.hstack([ozon[:, :s, :], ozon[:, e:, ]]).mean(axis=(1,2))
```

```python
df_time2 = pd.DataFrame({
    "time": time.astype('timedelta64[M]'),
    "o3_eq": average_eq(45),
    "o3_polar": average_polar(45),
})
df_time2.time = pd.Timestamp("1970-01-15T00:00:00").normalize() + df_time2.time
df_time2.set_index("time", inplace=True)

df_time2.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_time2.o3_eq, label='Equatorial')
plt.plot(df_time2.o3_polar, label='Polar')
plt.legend()
```

```python
df_timem = pd.merge_asof(df_time1, df_time2, on="time",
                         tolerance=pd.Timedelta(15, unit="D"),
                         direction='nearest')
df_timem.set_index("time", inplace=True)
df_timem.head()
```

```python
def plot_merged(df):
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax1.plot(df_timem.wolf, label='Wolf')
    ax2 = ax1.twinx()
    ax2.plot(df_timem.o3_eq, color='orange', label='O3 Eqatorial')
    ax2.plot(df_timem.o3_polar, color='gray', label='O3 Polar')
    ax1.legend()
    ax2.legend(loc="upper left")
```

```python
plot_merged(df_timem)
```

```python
df_timem.corr()
```
