## Временные ряды

```python
import datetime
import matplotlib.pyplot as plt
```

```python
parser = lambda date: datetime.datetime.strptime(date, '%d.%m.%Y %H:%M')
df_time1 = pd.read_csv('weather.csv.gz', sep=';', comment='#', index_col=None,
                 parse_dates=['Local time in Elista (airport)'], date_parser=parser)
```

```python
parser = lambda date: datetime.datetime.strptime(date, '%d.%m.%Y %H:%M') + pd.Timedelta('5 m')
df_time2 = pd.read_csv('UUEE.gz', sep=';', comment='#', index_col=None,
                 parse_dates=['Local time in Sheremetyevo (airport)'], date_parser=parser)
```

```python
plt.plot(df_time1['Local time in Elista (airport)'], df_time1['T'])
```

```python
df_time1
```

```python
df_time1['T'].hist(bins=50)
```

```python
new_df = pd.DataFrame()
```

```python
new_df['time'] = df_time1['Local time in Elista (airport)']
```

```python
new_df['T_El'] = df_time1['T']
```

```python
df_time2.rename(columns={'Local time in Sheremetyevo (airport)':'time', 'T':'T_M'}, inplace=True)
```

```python
new_df = new_df.sort_values(by=['time'])
df_time2 = df_time2.sort_values(by=['time'])
```

```python
df_time2.dtypes
```

```python
df_merged = pd.merge_asof(new_df, df_time2, on="time", tolerance=datetime.timedelta(minutes=10), direction='nearest')
```

```python
df_merged
```

## Другой пример для merge_asof

```python
d1 = {'x': np.arange(1, 100, 5),
     'x2': np.arange(1, 100, 5)**2}
```

```python
df1 = pd.DataFrame(d1)
```

```python
d2 = {'x': np.arange(2, 101, 5),
     'x2_shift': np.arange(2, 101, 5)**2}
```

```python
df2 = pd.DataFrame(d2)
```

```python
merged_df = pd.merge_asof(df1, df2, on='x', direction='nearest')
```
